```python
# pasting the previous video of people dictionary data

people = {
    "first": ['srinu','sow','abc'],
    "last" : ['balireddy','yalla','abc'],
    "email": ['abc@abc.com','kbc@kbc.com','abc@abc.com']
}

import pandas as pd
df = pd.DataFrame(people)
df
```

```python
# if we want to sort the df with last name, we can use sort_values method

df.sort_values(by='last' , ascending= False) # descending order
```

```python
# to sort multiple columns

df.sort_values(by=['last','first'], ascending = False)

# this will sort first on the last name and then the first name in descending order 
```

```python
# to sort last name in descending and first name in ascending order

df.sort_values(by=['last','first'], ascending=[False, True])
```

```python
# to set this sort permanent we need to set inplace= True

df.sort_values(by=['last','first'], ascending=[False, True], inplace=True)
df
```

```python
# to revert back the sort values and retain the initial index values
# we can sort using the index of the df

df.sort_index()
```

```python
# to sort a single column

df['last'].sort_values()
```

```python
# now lets go back to stackover flow dataset and manipulate the data using pandas.
# just consolidate the csv file reading code 


import pandas as pd
df        = pd.read_csv('data/survey_results_public.csv')
schema_df = pd.read_csv('data/survey_results_schema.csv')

pd.set_option('display.max_columns', 20)
pd.set_option('display.max_rows', 5)
```

```python
# sort the df by country name

df.sort_values(by='Country', inplace=True)
```

```python
df['Country'].head(50)
```

```python
# to check the salary and the country

df.sort_values(by=['Country','ConvertedComp'], ascending=[True, False], inplace=True)
```

```python
df[['Country','ConvertedComp']]
```

```python
# finding the largest and smallest values 
# if we want to see the 10 highest salary - nlargest() method

df['ConvertedComp'].nlargest()
```

```python
# in the above only the convertedComp is displayed, if we want to see the other columns as well

df.nlargest(10, 'ConvertedComp')  # first argument  is the no. of rows, second is column name
```

```python
# if we want to see the 10 highest salary - nsmallest() method

df['ConvertedComp'].nsmallest()
```

```python
# to display other columns

df.nsmallest(10, 'ConvertedComp')
```
