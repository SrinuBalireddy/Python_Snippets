```python
# pasting the previous video of people dictionary data

people = {
    "first": ['srinu','sow','abc'],
    "last" : ['balireddy','yalla','abc'],
    "email": ['abc@abc.com','kbc@kbc.com','abc@abc.com']
}

import pandas as pd
df = pd.DataFrame(people)
df
```

```python
# adding columns to the df
# simply create a column and pass in a series of values

# to combine first and last name to a new column fullname

df['first']+ ' ' + df['last']
```

```python
# assign it a new column

df['fullname'] = df['first']+ ' ' + df['last']
df

# we cannot use the dot notation while creating new columns.
```

```python
# removing columns
# we can use the drop methods

df.drop(columns=['first', 'last'])

# we have to use inplace= True for the changes to reflect in the df

df.drop(columns=['first', 'last'], inplace = True)
df
```

```python
# if we want to get back the first and last name columns 
# by splitting the fullname column, we can use str.split() method

df['fullname'].str.split()   # by default split will operate on space
```

```python
# if we want to assign the above results to new columns, we have to use EXPAND argument

df['fullname'].str.split(expand =True)
```

```python
# assigning the above results to new columns

df[['first', 'last']] = df['fullname'].str.split(expand=True)
```

```python
df
```

```python
""" Now adding rows to the df"""

# adding single row of data using APPEND method

df.append({'first':'Tony'})
```

```python
# we get the above error stating the df doesn't have an index
# so we can pass in ignore_index = True to the append method

df.append({'first':'Tony'}, ignore_index=True)

# remaining columns will have the value as NaN
```

```python
# for the above to result to effect the df, we have to use df = 

df = df.append({'first':'Tony'}, ignore_index=True)
df
```

```python
# to append an existing df to a new df

# create new df

people = {
    "first": ['srinu1','sow1'],
    "last" : ['balireddy1','yalla1'],
    "email": ['abc@abc.com1','kbc@kbc.com1']
}

import pandas as pd
new_df = pd.DataFrame(people)
new_df
```

```python
# appending new_df to df

df.append(new_df, ignore_index = True)
```

```python
# the above warning is because the sort =True , which will be gone during pandas 1 version

# unlike drop method there is no inplace= True for this

df = df.append(new_df, ignore_index = True, sort= True)
df

# below result is duplicated and it was executed multiple times
```

```python
""" removing the columns from df """

# removing a specific row by passing in index

df.drop(index = 4 , inplace = True)
```

```python
# deleting rows based on a conditional

filt = df['last'] == 'balireddy1'
df.drop(index=df[filt].index)   # here we are fetching the filt results indexes
```
